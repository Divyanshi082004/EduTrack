```python
import tensorflow as tf
```

```python
from tensorflow import keras
```

```python
import numpy as np
```

training data
house size(sq ft)


```python
X = np.array([500, 800, 1000, 1500, 2000], dtype=float)
```

Price (in lakhs)

```python
Y = np.array([20, 30, 40, 60, 80], dtype=float)
```

```python
model = keras.Sequential([
    keras.layers.Dense(10, activation='relu', input_shape=[1]),
    keras.layers.Dense(1)
])
```

```python
model.compile(
    optimizer='adam',
    loss='mean_squared_error'
)
```

```python
model.fit(X, Y, epochs=500, verbose=0)
```

```python
print(model.predict(np.array([[2500]])))
```
